```python
import pandas as pd
```

```python
doc_id = '1rQD9-dVIWu1Z4qrXwEZ_JJn4quMLItzMmj2WttzISs4'
url = f'https://docs.google.com/spreadsheets/d/{doc_id}/export?format=csv'
capitals_coords = pd.read_csv(url)
capitals_coords = capitals_coords.rename(columns={
    'Country Name': 'country',
    'Capital Name': 'capital',
    'Capital Latitude': 'latitude',
    'Capital Longitude': 'longitude'
})
capitals_coords
```

```python
capitals_coords = capitals_coords.replace('Macedonia', 'North Macedonia')
```

```python
capitals_coords[capitals_coords['Continent Name'] == 'Europe'] \
    [['country', 'latitude', 'longitude']].to_csv('coordinates.csv', index=False)
```
